# Gender wise analysis

```python
sns.countplot( x='Gender', data = df)
```

```python
ab = sns.countplot( x='Gender', data = df)

for bars in ab.containers:
    ab.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
```

```python
gender_wise_sales = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(x= 'Gender', y='Amount', data =gender_wise_sales)
```

# Age Wise Analysis

```python
df.columns
```

```python
sns.countplot( x='Age Group', data = df, hue = 'Gender')
```

```python
ab = sns.countplot( x='Age Group', data = df, hue = 'Gender')

for bars in ab.containers:
    ab.bar_label(bars)
```

```python
Age_wise_sales = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(x= 'Age Group', y='Amount', data =Age_wise_sales)
```

# State wise Analysis

```python
df.columns
```

```python
ab = sns.countplot( x='State', data = df, hue = 'Orders')

for bars in ab.containers:
    ab.bar_label(bars)
```

```python
State_wise_orders = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending = False)

sns.set(rc={'figure.figsize':(8,5)})
sns.barplot(x= 'State', y='Orders', data = State_wise_orders)
```

# Married Status wise analysis

```python
df.columns
```

```python
ab = sns.countplot( x='Married Status', data = df)

sns.set(rc={'figure.figsize':(7,5)})
for bars in ab.containers:
    ab.bar_label(bars)
```


# Occupation wise analysis

```python
sns.set(rc={'figure.figsize':(20,5)})
ab = sns.countplot( x='Occupation', data = df)

for bars in ab.containers:
    ab.bar_label(bars)
```

```python
Occupation_wise_orders = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'Occupation', y='Amount', data = Occupation_wise_orders)
```

# Product Category

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(25,5)})
ab = sns.countplot( x='Product_Category', data = df)

for bars in ab.containers:
    ab.bar_label(bars)
```

```python
Product_Category_wise_orders = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'Product_Category', y='Amount', data = Product_Category_wise_orders)
```

```python
Product_wise_orders = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'Product_ID', y='Orders', data = Product_wise_orders)
```
